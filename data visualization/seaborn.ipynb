```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('titanic')
df
```

```python
#countplot : survival based on gender 
sns.countplot(df, x='survived', )
plt.title('survived vs not survive by gender')
plt.show()
```

```python
sns.barplot(df, x='sex', y='fare',)
plt.title('average fare by passenger')
```

```python
sns.pointplot(df, x='class', y='survival by gender')
plt.title('survival rate by class')
plt.show()
```
